1. Using the mean squared error (MSE) as your objective function allows for a closed form solution to finding the maximum likelihood estimate (MLE) of your model parameters in linear regression. Let’s consider the simple, single predictor variable model $Y= \beta_0 + \beta_1 X $. 

a) Use algebra to show how you can expand out the MSE to get from i to ii below. Do your proof using Latex equation formatting.


 _i)_ $MSE(\beta_0, \beta_1)=E[ (Y-(\beta_0 + \beta_1 X))^2]$ 

**Answer:**

$= E[(Y-(\beta_0 + \beta_1X))(Y-(\beta_0+\beta_1X))] $

$= E[Y^2] - 2\beta_0E[Y] - 2\beta_1E[XY] + E[(\beta_0 + \beta_1 X)^2]$

$= E[Y^2] - 2\beta_0E[Y] - 2\beta_1(Cov[XY] + E[X]E[Y]) + E[(\beta_0 + \beta_1 X)(\beta_0 + \beta_1 X)]$

$= E[Y^2] - 2\beta_0E[Y] - 2\beta_1(Cov[XY] + E[X]E[Y]) + \beta_0^2 + 2\beta_1E[X] + \beta_1^2E[X^2]$


_ii)_ $=E[Y^2] -2 \beta_0E[Y]-2 \beta_1 Cov[X,Y]-2 \beta_1 E[X]E[Y]+ \beta_0^2 +2 \beta_0 \beta_1 E[X]+\beta_1^2 Var[X]+ \beta_1^2 (E[X])^2$

b) Prove that the MLE of $\beta_0$ is $E[Y]- \beta_1 E[X]$ by taking the derivative of _ii_ above, with respect to $\beta_0$, setting the derivative to zero, and solving for $\beta_0$.

**Answer:**

$\frac{\delta E[(Y-(\beta_0 + \beta_1X))^2]}{\beta_0} = -2E[Y] + 2\beta_0 + 2\beta_1E[X]$

$ 0 = -2E[Y] + 2\beta_0 + 2\beta_1E[X]$

$ -2\beta_0 = -2E[Y] + 2\beta_1E[X] $

$ \beta_0 = E[Y] - \beta_1E[X] $
      

c) Prove that the MLE for $\beta_1$ is $Cov[X,Y]/Var[X]$ by taking the derivative of _ii_ above, with respect to $\beta_1$, setting the derivative to zero, and solving for $\beta_1$. 

**Answer:**

$\frac{\delta E[(Y-(\beta_0 + \beta_1X))^2]}{\beta_1} = -2Cov[XY] -2E[X]E[Y] + 2\beta_0E[X] + 2\beta_1Var[X] + 2\beta_1(E[X])^2$

$ 0 =  -2Cov[XY] -2E[X]E[Y] + 2\beta_0E[X] + 2\beta_1Var[X] + 2\beta_1(E[X])^2 $

$ - 2\beta_1Var[X] - 2\beta_1(E[X])^2 =  -2Cov[XY] -2E[X]E[Y] + 2\beta_0E[X]  $

$ - 2\beta_1Var[X] - 2\beta_1(E[X])^2 =  -2Cov[XY] -2E[X]E[Y] + 2(E[Y] - \beta_1E[X])E[X]  $

$ - 2\beta_1Var[X] - 2\beta_1(E[X])^2 =  -2Cov[XY] -2E[X]E[Y] + 2E[X]E[Y] - 2\beta_1E[X]^2  $

$ - 2\beta_1Var[X] - 2\beta_1(E[X])^2 =  -2Cov[XY] - 2\beta_1E[X]^2  $

$ - 2\beta_1Var[X]  =  -2Cov[XY] $

$ \beta_1 = Cov[XY]/Var[X] $

4. Three methods for evaluating a regression model’s performance are $r^2$, $RSE$, and the $F$ Test. Show the functional form of each method and describe conceptually what it is telling you about your model.

**Answer:**

* **R-square:** 

$r^2=\frac{(TSS-RSS)}{TSS}=1-(\frac{RSS}{TSS})= 1 - \frac{\Sigma_{i=1}^n (y_i - \hat{y_i})^2}{\Sigma_{i=1}^n (y_i - \bar{y_i})^2}$

The $r^2$ tells you the percentage of the variance in the response variable $Y$ that can be explained by the model.

* **Residual standard error (RSE): **

$RSE= \sqrt{\frac{\Sigma_{i=1}^n (y_i - \hat{y_i})^2}{(n-2)}} = \sqrt{\frac{RSS}{n-2}}$

The RSE provides a measure of confidence of the model fit. Smaller values of RSE indicate greater confidence that your model explains the data.

* **F Test (F):**

$F=\frac{(TSS-RSS)p^{-1}}{RSS (n-p-1)^{-1}} = \frac{ (\Sigma_{i=1}^n (y_i - \bar{y_i})^2) - (\Sigma_{i=1}^n (y_i - \hat{y_i})^2) p^{-1}}{\Sigma_{i=1}^n (y_i - \hat{y_i})^2 (n-p-1)^{-1}} $ 


If the variance not explained by your model (TSS) is greater than the variance explained your model (RSS), then the ratio is high and F is significant

6. What are high leverage observations? How do they impact your model fits? What assumption of linear regression do they violate?

**Answer:**

High leverage observations are outliers that, because of their deviation from the rest of the distribution, have a much stronger influence on the estimated model fit than the rest of the observations. These tend to pull the model fit away from the true maximum likelihood estimate. They violate the assumption that your samples are independent and identically distributed.

7. A linear mixed effect model has the form $Y = X\beta + Z\upsilon + \epsilon$. 

a) What are the assumptions about $Z$ and $\upsilon$ that make this different from using a regular control variable in a simple linear regression model?


**Answer:**

For linear mixed effects models we assume that Z is a categorical variable whose values are arbitrary (e.g., batch ID, subject ID). We also assume that is random as well (i.e., $\upsilon \sim N(\mu,\sigma)$). When using a control variable in simple linear regression, you assume that the regression coefficient for that variable is not random but meaningful in its relationship to the response variable. You also assume in simple regression that the value of the regressor itself maps to a meaningful relationship in the data (e.g., gender, age, education). 


b) What is the objective function for this model (i.e., the function that needs to be minimized to find the optimal parameters) and how is it different than the objective function for the ordinary least squares model?


**Answer:**

Linear Mixed Effects: $min{||Y-X\beta-Z\Lambda_{\theta}\upsilon||^2+||\upsilon||2}$

Linear Regression: $min{||Y-X\beta||^2}$

The linear mixed effects objective function includes the covariance matrix of the random effect term itself ($\Lambda_{\theta}$) and adds an additional penalty for low values of $\upsilon$ (i.e., regularizes the fits on the random effects). Whereas the linear regression objective function only attempts to minimize the mean squared error.


c) What advantage does including a random effects term in your model give you with regards to understanding the fixed effects relationship?

**Answer:**

It allows for you to account for predictor variables that are not of primary interest and result from random selection, but may still impact the response variable itself.
